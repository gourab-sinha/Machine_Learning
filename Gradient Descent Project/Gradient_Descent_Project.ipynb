```
import numpy as np
```

```
class Linear_Regression_Model:
  def predict(self, X_test, m_slope):
    row,col = X_test.shape
    constants = np.ones([row,1],dtype=int)
    X_test = np.concatenate((X_test,constants),axis=1)
    Y_predict = np.zeros(row)
    for i in range(row):
      for j in range(len(m_slope)):
        Y_predict[i] += m_slope[j]*X_test[i][j]

    return Y_predict

  def score(self,Y_test,Y_pred):
    u = ((Y_test-Y_pred)**2).sum()
    v = ((Y_test-Y_test.mean())**2).sum()
    return 1-u/v

class Gradient_Descent:
  def __cost(self, pointsx,pointsy,m_slope):
    cost = 0
    N = len(pointsx)
    for i in range(N):
      m_x = pointsx[i]*m_slope
      cost += (1/N)*((pointsy[i]-sum(m_x))**2)

    return cost



  def __step_grad(self, pointsx, pointsy, learning_rate, m_slope):
    new_slope = np.zeros(len(m_slope))
    # print(new_slope)
    N = len(pointsx)
    for i in range(N):
      m_x = pointsx[i]*m_slope
      # print(m_x)
      for j in range(len(pointsx[i])):
        new_slope[j] += (-2/N)*(pointsy[i]-m_x.sum())*pointsx[i][j]

    m_new_slope = np.zeros(len(m_slope))
    for i in range(len(m_slope)):
      m_new_slope[i] = m_slope[i] - learning_rate*new_slope[i]

    return m_new_slope




  def grad(self, pointsx, pointsy, learning_rate, iteration):
    m_slope = np.zeros(len(pointsx[0]))
    m_slope[len(pointsx[0])-1] = 1

    for i in range(iteration):
      m_slope = self.__step_grad(pointsx,pointsy,learning_rate,m_slope)
      print("Cost: ", self.__cost(pointsx,pointsy,m_slope))

    return m_slope
```
